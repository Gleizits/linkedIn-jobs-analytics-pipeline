# ETL PROCESS

## Extract

```python
def extract():
  # Try y Except por si ocurren exepciones durante el proceso
  try:
        # Intenta cargar el archivo
        df = pd.read_csv(
            'job_skills.csv',
            sep='\t',
            engine='python',
            encoding='utf-8',
            on_bad_lines='skip'
        )

        print(f"Filas cargadas    : {len(df):,}")
        print(f"Columnas          : {list(df.columns)}")
        print(f"Tipos de datos    :\n{df.dtypes}")
        print(f"Pequeña muestra   :\n{df.head(3)}")

        return df

  except FileNotFoundError:
        print("El archivo no se encontró.")
        return None

  except Exception as e:
        print(f"Ocurrió un error al cargar el archivo: {e}")
        return None

extract()
```

# Transform

```python
def transform(df):
    # Renombrar columnas si vienen sin header
    df.columns = ['job_link', 'job_skills']

    # Eliminar filas duplicadas
    antes = len(df)
    df.drop_duplicates(subset='job_link', inplace=True)
    print(f"Duplicados eliminados : {antes - len(df):,}")

    # Eliminar nulos
    df.dropna(subset=['job_skills'], inplace=True)
    print(f"Filas limpias         : {len(df):,}")

    # Limpiar espacios y comillas en skills
    df['job_skills'] = df['job_skills'].str.strip().str.strip('"')

    # Una fila por skill
    df_expanded = df.assign(
        skill=df['job_skills'].str.split(',')
    ).explode('skill')

    df_expanded['skill'] = df_expanded['skill'].str.strip()
    df_expanded = df_expanded[df_expanded['skill'] != '']

    print(f"Total skill-filas     : {len(df_expanded):,}")
    print(f"Skills únicas         : {df_expanded['skill'].nunique():,}")
    print(f"Muestra:\n{df_expanded.head(5)}")

    return df_expanded

  transform(df)
```
